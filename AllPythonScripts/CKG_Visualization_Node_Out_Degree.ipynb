# Node out Degree Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from collections import Counter
import csv
```

```python
file_path = 'extracted_diseases.csv'

# Initialize a list to store the numbers
numbers = []

with open(file_path, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  

    for row in reader:
        # Ensure there's a second column
        if len(row) >= 2:
            try:
                # Convert the second column value to integer and append to the list
                number = int(row[1].strip())
                numbers.append(number)
            except ValueError:
                continue 
```

```python
# Count the occurrences of each number
number_counts = Counter(numbers)
#number_counts
```

```python
# Extract the unique numbers and their counts
node_degree = sorted(number_counts.keys())
node_degree_freq = [number_counts[num] for num in node_degree]
```

```python
print(node_degree)
#print(node_degree_freq)
```

```python
# checking is there are any 0 values on x and y values
zero_values = [x for x in node_degree_freq if x == 0]
print("Values equal to 0:", zero_values)
```

```python
# Plotting the graph
plt.figure(figsize=(10, 6))
plt.semilogx(node_degree, node_degree_freq, marker='o', linestyle='-', color='b')
plt.title('Node Degree Distribution')
plt.xlabel('node degree')
plt.ylabel('# of nodes for a given in/out degree')

plt.grid(True)
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```

```python
# plotting with Y-axis log scale
# Plotting the graph
plt.figure(figsize=(10, 6))
plt.semilogx(node_degree, node_degree_freq, marker='o', linestyle='-', color='b')
plt.yscale('log')
plt.title('Node Degree Distribution')
plt.xlabel('node degree')
plt.ylabel('# of nodes for a given in/out degree')

plt.grid(True)
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```

```python
node_degree = node_degree[:10]
node_degree_freq = node_degree_freq[:10]
```

```python
# Trying different binning technique 
# Binning the node counts
bins = np.logspace(np.log10(min(node_degree)), np.log10(max(node_degree)), 50)

# Creating the histogram
plt.hist(node_degree, bins=bins, weights=node_degree_freq, alpha=0.7, edgecolor='black')

# Applying a logarithmic scale to the y-axis
plt.yscale('log')

# Adding labels and title
plt.xlabel('node degree')
plt.ylabel('# of nodes for a given in/out degree')
plt.title('Histogram - Node Degree Distribution')

# Displaying the plot
plt.grid(True)
plt.show()
```

```python
# Determine the number of bins (replace 50 with your desired number of bins)
num_bins = 10

# Compute histogram values and bin edges
hist_values, bin_edges = np.histogram(node_degree, bins=num_bins)

# Plot the histogram
plt.hist(node_degree, bins=bin_edges, edgecolor='black')
plt.yscale('log')

# Add labels and title
plt.xlabel('node degree')
plt.ylabel('# of nodes for a given in/out degree')
plt.title('Histogram - Node Degree Distribution')

# Display the plot
plt.grid(True)
plt.show()
```
