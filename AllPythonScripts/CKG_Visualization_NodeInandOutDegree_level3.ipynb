```python
import os
import torch
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from torch_geometric.utils import degree
```

```python
folder_path = 'data 1/'
```

```python
folder_path = 'data 1/'  

for file_name in os.listdir(folder_path):
    full_file_path = os.path.join(folder_path, file_name)
    edge_index = pd.read_csv(full_file_path).values.T
    edge_index = torch.tensor(edge_index, dtype=torch.long)
    
    # Calculate out-degree and in-degree
    out_degree = degree(edge_index[0])
    in_degree = degree(edge_index[1])
    parts = file_name[:-4].split("_") 
    source_type = parts[0]
    edge_label = "_".join(parts[1:-1])  # Join back the middle parts
    target_type = parts[-1]
   
      print(f"File: {file_name}, Source: {source_type}, Edge: {edge_label}, Target: {target_type}")
#     print(f"Out-degree: {out_degree}, In-degree: {in_degree}")
#     plt.figure(figsize=(10, 6))
#     plt.hist(out_degree.numpy(), bins=range(int(out_degree.max().item()) + 1), color='skyblue', edgecolor='black')
#     plt.title(f" Source: {source_type}, Edge: {edge_label}, Target: {target_type}  Out Degree")
#     plt.xlabel('Out-Degree')
#     plt.ylabel('Frequency')
#     plt.xticks(range(int(out_degree.max().item()) + 1))
#     plt.grid(axis='y', alpha=0.75)
#     plt.show()
 
    
```

```python
 # Binning the node counts
#bins = np.logspace(np.log10(min(out_degree)), np.log10(max(out_degree)), 50)
bins=range(int(out_degree.max().item()) + 1)
plt.figure(figsize=(10, 6))
plt.hist(out_degree.numpy(), bins= bins, color='skyblue', edgecolor='black')
plt.title(f" Source: {source_type}, Edge: {edge_label}, Target: {target_type}  Out Degree")
plt.xlabel('Out-Degree')
plt.ylabel('Frequency')
plt.xticks(range(int(out_degree.max().item()) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
 
```

```python
plt.figure(figsize=(10, 6))
plt.hist(out_degree.numpy(), bins=range(int(in_degree.max().item()) + 1), color='skyblue', edgecolor='black')
plt.title('Histogram of Node Out-Degrees')
plt.xlabel('I-Degree')
plt.ylabel('Frequency')
plt.xticks(range(int(in_degree.max().item()) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
 
```
