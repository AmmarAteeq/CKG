# Relationship type and count EDA

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the data
data_path = 'NumberOfEdges.csv'
data = pd.read_csv(data_path)
```

```python
# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Set the figure size for better readability
plt.figure(figsize=(12, 8))
```

```python
# Bar Chart
plt.subplot(1, 2, 1)
plt.barh(data['RelationshipType'], data['RelationshipCount'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Relationship Type')
plt.title('Relationship Count by Type')
```

```python

plt.tight_layout()  # Adjust layout to not overlap
plt.show()
```

```python
# printing all counting
data.sort_values(by='RelationshipCount', ascending=False)
```

```python
data.count()
```

```python
# display the top 10 and bottom 10 relationship types, we'll sort the dataframe by 'RelationshipCount' in descending order
# and then select the top 10 rows.

top_10_data = data.sort_values(by='RelationshipCount', ascending=False).head(10)
bottom_10_data = data.sort_values(by='RelationshipCount', ascending=True).head(10)



# Set the figure size for a clearer view
plt.figure(figsize=(10, 6))

# Bar Chart for Top 10 Relationship Types
plt.barh(top_10_data['RelationshipType'], top_10_data['RelationshipCount'], color='lightgreen')
plt.xlabel('Count')
plt.ylabel('Relationship Type')
plt.title('Top 10 Relationship Types by Count')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top
plt.show()

# Displaying the top 10 data in tabular format as well
top_10_data
```

```python
# Bar Chart for Top 10 Relationship Types
plt.barh(bottom_10_data['RelationshipType'], bottom_10_data['RelationshipCount'], color='lightgreen')
plt.xlabel('Count')
plt.ylabel('Relationship Type')
plt.title('Bottom 10 Relationship Types by Count')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest count at the top
plt.show()
# Displaying the top 10 data in tabular format as well
bottom_10_data
```

```python
# Pie Chart for Top 10 Relationship Types
plt.figure(figsize=(8, 8))
plt.pie(top_10_data['RelationshipCount'], labels=top_10_data['RelationshipType'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Relationship Types Proportions')
plt.show()
```

```python
# Pie Chart for Bottom 10 Relationship Types
plt.figure(figsize=(8, 8))
plt.pie(bottom_10_data['RelationshipCount'], labels=bottom_10_data['RelationshipType'], autopct='%1.1f%%', startangle=140)
plt.title('Bottom 10 Relationship Types Proportions')
plt.show()
```

```python
# Calculate the cumulative distribution function (CDF)
data_sorted = data['RelationshipCount'].sort_values()
cdf = np.arange(1, len(data_sorted)+1) / len(data_sorted)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(cdf,data_sorted,marker='.', linestyle='none')
plt.xlabel('Relationship Count')
plt.ylabel('CDF')
plt.yscale('log')
plt.title('Cumulative Distribution Function of Relationship Counts')
plt.grid(True)



plt.show()
```

```python
# To make the x-axis values more readable, especially for the top frequency graphs, 
# we'll adjust the histogram to focus on the range with the highest frequencies.
# We aim to make the x-axis values for the top 4 frequency graphs readable.

# Find the range that includes the top frequencies
# We'll use a smaller number of bins to focus on the most frequent counts and adjust the xlim to make the x-axis values readable

# First, let's determine an appropriate range for the x-axis by examining the data
max_count_for_top_frequencies = data['RelationshipCount'].quantile(0.95)  # A rough estimate to focus on the most common counts

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data['RelationshipCount'], bins=30, color='blue', edgecolor='black')

# Find the bins corresponding to the top 4 frequencies to adjust the x-axis range accordingly
top_bins = sorted(zip(n, bins[:-1]), reverse=True)[:4]  # Sort by frequency and take top 4
min_bin = min(top_bins, key=lambda x: x[1])[1]
max_bin = max_count_for_top_frequencies  # Use the 95th percentile as an upper limit for readability

plt.xlabel('Relationship Count')
plt.ylabel('Frequency')
plt.title('Histogram of Relationship Counts (Focused Range)')
plt.grid(True)

# Adjust the x-axis to focus on the most frequent counts, enhancing readability
plt.xlim(min_bin, max_bin)

# Change the x-axis to plain style to avoid exponent values and improve readability
plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.show()
```
